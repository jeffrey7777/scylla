```python
# cleaning data
# Changing the columns names
# previewing the column names

dataset1 = df1.set_axis(['PRODUCT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_ID'],  axis=1, inplace=False)
dataset1.head()
```

```python
#mapping and applying the same to all datasets
dataset2.columns = dataset1.columns
dataset2.head()
```

```python
dataset3.columns = dataset1.columns
dataset3.head()
```

```python
# combining the three datasets using concatenation

df_combine = pd.concat([dataset1,dataset2,dataset3], ignore_index = 1)
df_combine.tail()
```

```python
# Dropping unnecessary columns 

df_combine_cleaned = df_combine.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE'])
df_combine_cleaned.head()
```
